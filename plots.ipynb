```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
method = ['Resize (128x128)', '(100x100,0 padding)', '(100x100,14 padding)', '(200x200,28 padding)', '(400x400,0 padding)']
results = [0.774, 0.853, 0.890, 0.902, 0.895]
```

```python
def autolabel(rects, pop_growth):
    for idx, rect in enumerate(rects):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2, height,
                '{}'.format(pop_growth[idx]),
                ha='center', va='bottom', rotation=0)

```

```python
p1 = plt.bar(method, results)
autolabel(p1, results)
plt.ylim([0.75, 0.925])
plt.ylabel('F1-Score')
plt.xticks(rotation=-30)
plt.title('F1-Score on test-set (AICrowd)')

plt.savefig('./litterature/F1_scores.png', bbox_inches = 'tight')
plt.show()
```

```python
val_f1 = pd.read_csv('plots/valf1-best.csv')
f1 = pd.read_csv('plots/f1-best.csv')
max_f1 = val_f1['Value'].max()
```

```python
#plt.figure(figsize = (10,5))
plt.plot(f1.Step.values + 1,f1.Value.values, label= 'F1 Score - Training', color = 'r')
plt.plot(val_f1.Step.values+ 1, val_f1.Value.values, label= 'F1 Score - Validation', color = 'b')
plt.scatter(30, max_f1, color='b')
plt.text(24.5, (max_f1-0.018), 'F1 = 0.985', fontsize = 12)

plt.title('Training - F1 Scores')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')

plt.legend()
plt.savefig('plots/f1-training.png')
```
