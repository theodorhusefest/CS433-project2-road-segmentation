```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
method = ['Resize (128x128)', '(100x100,0 padding)', '(100x100,14 padding)', '(200x200,28 padding)', '(400x400,0 padding)']
results = [0.77, 0.85, 0.89, 0.90, 0.895]
```

```python
plt.bar(method, results)
plt.ylim([0.75, 0.925])
plt.ylabel('F1-Score')
plt.xticks(rotation=-30)
plt.title('F1-Score on test-set (AICrowd)')
plt.savefig('./litterature/F1_scores.png', bbox_inches = 'tight')
plt.show()
```
